```python
# Import required libraries and dependencies
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame and make the index the "coin_id" column.
market_data_df = pd.read_csv("Resources/crypto_market_data.csv", index_col="coin_id")

# Display sample data
market_data_df.head(10)
```

```python
# Generate summary statistics
market_data_df.describe()
```

```python
market_data_df.dtypes
```

```python
market_data_df.columns
```

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
market_data_scaled = StandardScaler().fit_transform(market_data_df)
```

```python
# Create a DataFrame with the scaled data
df_market_data_transformed = pd.DataFrame(market_data_scaled, 
                                          columns=['price_change_percentage_24h', 'price_change_percentage_7d',
                                                   'price_change_percentage_14d', 'price_change_percentage_30d',
                                                   'price_change_percentage_60d', 'price_change_percentage_200d',
                                                   'price_change_percentage_1y'])

# Copy the crypto names from the original data
df_market_data_transformed['coin_id'] = market_data_df.index

# Set the coinid column as index
df_market_data_transformed.set_index('coin_id', inplace=True)

# Display sample data
display(df_market_data_transformed.head())
```

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values to try
# Use a range from 1 to 11
k = list(range(1, 11))

# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
for i in k:
    Model = KMeans(n_clusters=i, random_state=1)
    Model.fit(df_market_data_transformed)
    inertia.append(Model.inertia_)

# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using the scaled DataFrame
# 3. Append the model.inertia_ to the inertia list
# Answer is below the "create for loop" and above "inside the loop"


# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    "k": k,
    "inertia": inertia
}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow_curve = pd.DataFrame(elbow_data)

# Display the DataFrame
display(df_elbow_curve.head(10))
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
df_elbow_curve.plot.line(
    x="k",
    y="inertia",
    title="Elbow Curve",
    xticks=k,
    xlabel="Number of k clusters",
    ylabel="Inertia"
)
```

#### Answer the following question: 
**Question:** What is the best value for `k`?

**Answer:** Based on the analysis of the graph, the most optimal value is number 4 due to it being the steepest drop from the 2nd and 3rd k clusters.

### Cluster Cryptocurrencies with K-means Using the Original Scaled Data.

```python
# Initialize the K-Means model using the best value for k
Model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the scaled data
Model.fit(df_market_data_transformed)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
crypto_clusters = Model.predict(df_market_data_transformed)

# View the resulting array of cluster values.
display(crypto_clusters)
```

```python
# Create a copy of the DataFrame
df_clustered_market_data = df_market_data_transformed.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
df_clustered_market_data["crypto_clusters"] = crypto_clusters

# Display sample data
display(df_clustered_market_data.head())
```

```python
# Create a scatter plot using Pandas plot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Use "rainbow" for the color to better visualize the data.
df_clustered_market_data.plot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    c="crypto_clusters",
    colormap="rainbow")
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` on the original scaled DataFrame to reduce to three principal components.
pca_scaled_model = pca.fit_transform(df_market_data_transformed)

# View the first five rows of the DataFrame. 
pca_scaled_model[0:5]
```

```python
# Retrieve the explained variance to determine how much information can be attributed to each principal component.
pca.explained_variance_ratio_
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The total explained variance of the combined principal components is 0.89503166.

```python
# Create a new DataFrame with the PCA data.
# Note: The code for this step is provided for you

# Creating a DataFrame with the PCA data
df_pca_data = pd.DataFrame(pca_scaled_model,
                           columns=["PCA1", "PCA2", "PCA3"])

# Copy the crypto names from the original data
df_pca_data["coin_id"] = market_data_df.index

# Set the coinid column as index
df_pca_data.set_index("coin_id", inplace=True)

# Display sample data
display(df_pca_data.head())
```

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values to try
# Use a range from 1 to 11
k_values = list(range(1, 11))

# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using PCA DataFrame.
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_means_model = KMeans(n_clusters=i, random_state=1)
    k_means_model.fit(df_pca_data)
    inertia.append(k_means_model.inertia_)


# Create a dictionary with the data to plot the Elbow curve
elbow_curve_data = {
    "k_values":k_values,
    "inertia":inertia
}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow_method = pd.DataFrame(elbow_curve_data)

# Display the DataFrame
display(df_elbow_method)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
df_elbow_method.plot.line(
    x="k_values",
    y="inertia",
    title="Elbow Curve",
    xticks=k_values,
    xlabel="Number of k clusters",
    ylabel="Inertia"
)
```

#### Answer the following questions: 
* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** The Best value for k, as seen on the graph, is 4.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** Absolutely not, they are exactly the same.

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
k_means_model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the PCA data
k_means_model.fit(df_pca_data)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
crypto_cluster = k_means_model.predict(df_pca_data)

# Print the resulting array of cluster values.
crypto_cluster
```

```python
# Create a copy of the DataFrame with the PCA data
df_clustered_pca_data = df_pca_data.copy()

# Add a new column to the DataFrame with the predicted clusters
df_clustered_pca_data["crypto_cluster"] = crypto_cluster

# Display sample data
display(df_clustered_pca_data.head())
```

```python
# Create a scatter plot using hvPlot by setting `x="PCA1"` and `y="PCA2"`. 
df_clustered_pca_data.plot.scatter(
    x="PCA1",
    y="PCA2",
    c="crypto_cluster",
    colormap="winter")
```

#### What is the impact of using fewer features to cluster the data using K-Means?
* **Answer:** The impact of using fewer features procures dimensionality reduction and computational efficiency. It helps to simplify the model and reduce noise by utilizing less memory to store complex datasets with limited hardware. Convergence speed and scalability offer faster coverage with increased k_cluster computation times. 

### Determine the Weights of Each Feature on each Principal Component

```python
# Use the columns from the original scaled DataFrame as the index.
pca_component_weights = pd.DataFrame(pca.components_, columns=df_market_data_transformed.columns, index=["PCA1", "PCA2", "PCA3"])

display(pca_component_weights)
```
